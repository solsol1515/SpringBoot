```python
%matplotlib inline
    
import matplotlib.pyplot as plt
import numpy as np

""" 0. 플롯 스타일 """
plt.style.use('classic')
#plt.style.use('seaborn-whitegrid')

x=np.linspace(0, 10, 100)  # 0~10 사이의 수를 100개를 등균하게?
# x
plt.plot(np.sin(x))
```

```python
""" 1. 색상지정 """
plt.plot(np.sin(x-0), color='blue') # 색상이름 지정
plt.plot(np.sin(x-1), color='g')    # 색상코드명
plt.plot(np.sin(x-2), color='#FFDD44') # 16진수 RGB
plt.plot(np.sin(x-3), color='chartreuse') # HTML 색상이름
plt.plot(np.sin(x-4), color='0.75') # 0과 1사이의 회색조 ( 0: 검정)
plt.plot(np.sin(x-5), color=(1.0, 0.2, 0.5)) # RGB 튜플형식(0과 1사이)
```

```python
""" 2. 선스타일 """

# plt.plot(np.sin(x-0), linestyle='solid')
# plt.plot(np.sin(x-1), linestyle='dashed')
# plt.plot(np.sin(x-2), linestyle='dashdot')
# plt.plot(np.sin(x-3), linestyle='dotted')

plt.plot(np.sin(x-0), linestyle='-') # solid
plt.plot(np.sin(x-1), linestyle='--') # dashed
plt.plot(np.sin(x-2), linestyle='-.') # dashdot
plt.plot(np.sin(x-3), linestyle=':') # dotted
```

## r(red) g(green) b(blue) c(cyan) m(magenta) y(yellow) k(black) 

```python
""" 3. 선과 스타일을 간결하게 (대략 읽는 수준으로 ) """

plt.plot(np.sin(x-0), '-g')  # solid + green  '-g'  'g-' 동일
plt.plot(np.sin(x-1), '--c') # dashed + cyan
plt.plot(np.sin(x-2), '-.k') # dashdot + black
plt.plot(np.sin(x-3), ':r')  # dotted + red

```

```python
''' 4. 범례 ( 위와 동일하고 label 추가 )'''

# plt.figure()

plt.plot(np.sin(x-0), '-g' , label='first')  
plt.plot(np.sin(x-1), '--c', label='second') 
plt.plot(np.sin(x-2), '-.k', label='third')
plt.plot(np.sin(x-3), ':r',  label='fourth')  

plt.grid()

# 범례 : loc=위치인데 순번은?   오른쪽 상단이 1로,  왼쪽 상단이 2, 왼쪽 하단이 3, 오른쪽 하단이 4
plt.legend(loc=2)

# plt.show()
```
