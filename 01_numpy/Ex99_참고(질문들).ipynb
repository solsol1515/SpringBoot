**역행렬**

**원래 행렬과 역핵렬을 행렬의 곱으로 하면 단위행렬**

```python
import numpy as np

x = np.array([[1,2],[3,4]])

# 역행렬
y = np.linalg.inv(x)
y
```

```python
# 행렬의 곱

np.dot(x,y)
```

```python
np.matmul(x,y)
```

```python
x = np.array([[1,2,3],[3,4,5],[4,5,6]])

# 역행렬
y = np.linalg.inv(x)
y

# 행렬의곱
print(np.dot(x,y))
print(np.matmul(x,y))
```

```python
x = np.array([[1,2,3,4],[3,4,5,6],[4,5,6,6],[5,6,7,8]])

# 역행렬
y = np.linalg.inv(x)
y

# 행렬의곱
print(np.dot(x,y))
print(np.matmul(x,y))
```

```python
a = np.ones([9, 5, 7, 4])
c = np.ones([9, 5, 4, 3])

print(a)
print('-'*50)
print(c)
print('='*50)

print(np.dot(a, c).shape) #  (9, 5, 7, 9, 5, 3)
print(np.matmul(a, c).shape) #  (9, 5, 7, 3)
# n is 7, k is 4, m is 3
```
